```python
generos = data.genres
generos
```

```python
lista_generos_data_w_genres = []
num_generos_contidos = []

for genero in generos:
    genero = genero.replace('[','').replace(']','').replace('"','').replace("'","").split(',')
    num_generos_contidos.append(len(genero))
    for gen in genero:
        gen = gen.strip()
        lista_generos_data_w_genres.append(gen)
len(lista_generos_data_w_genres)
```

```python
data_generos = pd.DataFrame({'Gênero': lista_generos_data_w_genres})
data_generos.shape
```

```python
top_10_generos = data_generos['Gênero'].value_counts()[:10]
```

```python
top_10_generos.plot(kind='bar')
```
